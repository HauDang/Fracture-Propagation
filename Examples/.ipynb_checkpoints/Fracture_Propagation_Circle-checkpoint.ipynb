# Fracture Propagation in a Circle

This example demonstrate how to simulate a single fracture propagate.

```python
from IPython.display import Image
Image(filename="ModelCircle.jpg", width = 340, height = 550)
```

Starting by importing some modules

```python
from IPython import get_ipython
get_ipython().magic('reset -sf') 
import numpy as np
import Grid
import AuxiliaryFunction
import FEM
import matplotlib.pyplot as plt
```

Material parameters

```python
mt = Grid.MeshTreatment([],[])
mp = Grid.MatlabPlot()
af = AuxiliaryFunction.AuxiFu()
discre = FEM.Discretization()
boucon = FEM.SetupModel()
solution = FEM.PostProcessing()
thick = 25E-3
young = 15*10**9
poisson = 0.21
material = young/(1-poisson**2)*np.array(([[1,poisson,0],
                                    [poisson,1,0],
                                    [0,0,(1-poisson)/2]]))*thick
k = (3 - poisson)/(1 + poisson)
KIC = 2*1E6
fy = 1E6
fy0 = 0.005*fy
```

Define a grid

```python
p, t = mt.gmshpy(open('Mesh_circle_45.msh'))
p, t, nodaro, tipcra, nodcra, iniang, R, dc, aremin, ladv = mt.reprocessing()
numtip = len(tipcra)
p6, t6, qpe = mt.t3tot6(p,t,tipcra)
```

Before runing

```python
gloang = np.zeros((numtip))    
crapro = np.array(([True, True]))
error = 0
proste = 0
keq = np.zeros((2))
remesh = 1
dsi = np.zeros((numtip)) + ladv*4
rm = Grid.ReMesh( p, t, tipcra, nodcra, nodaro, dc, ladv )
```

## Main program

```python
data = []
while min(dsi) > ladv*3 and proste < 25:
    if remesh == 1:
        if proste > 1:
            p, t = rm.refinement( crapro, error, aremin, 'based on error' )
        p, t, moucra, roscra = rm.quaterelement( )
        p6, t6, qpe = mt.t3tot6(p,t,tipcra)
        lefhs = discre.stiffness( p6, t6, material, qpe) 
    dirdof, dirval, neudof, neuval = boucon.circle( p6, t6, fy )
    righs = discre.loadpoint( p6, t6, thick, neudof, neuval )
    disp = solution.solver(lefhs,righs,dirdof,dirval)
    ki, keq, craang = solution.SIF(p6, t6, numtip, disp, young, poisson, k, qpe )
    if max(abs(keq)) >= KIC:
        print('# propagation step =',proste, '# load =',fy, '# Keq =',keq)
        crapro = abs(keq) >=  KIC
        proste = proste + 1; remesh = 1
        p, t, tipcra, nodcra, moucra, gloang = rm.splitelement( crapro, moucra, roscra, iniang, gloang, craang )
        data.append(np.array(([np.copy(p)],[np.copy(t)],[np.copy(nodcra)],[np.copy(tipcra)],[np.copy(proste)],[np.copy(fy)])))
        Sigxxi,Sigyyi,Sigxyi,error = solution.stresses(p6, t6, disp, material, qpe)
        for i in range(numtip):
            dsi[i] = af.p2segment(tipcra[i], nodaro)
    else:
        remesh = 0
        fy = fy + fy0
```

```python
import matplotlib.pyplot as plt
for i in range(0,len(data),2):
    datai = data[i]
    fig, grid = plt.subplots()    
    mp.plot(fig, grid, datai[0][0],datai[1][0],'b-',datai[2][0],'r-',datai[3][0],'r', datai[4][0], datai[5][0])
    plt.pause(0.0001)  
```
