# Adaptive remeshing technique - Applied for fracture mechanis

This example demonstrate how to do an adaptive remeshing based on an error estimator for fracture problem.

```python
from IPython.display import Image
Image(filename="ModeEdgeCrack.jpg", width = 340, height = 550)
```

Starting by importing some modules

```python
from IPython import get_ipython
get_ipython().magic('reset -sf') 
import numpy as np
import math
import Grid
import AuxiliaryFunction
import FEM
from matplotlib.animation import FuncAnimation
import matplotlib.pyplot as plt
```

Material parameters

```python
sm = Grid.SimpleMesh()
mt = Grid.MeshTreatment([],[])
mp = Grid.MatlabPlot()
af = AuxiliaryFunction.AuxiFu()
discre = FEM.Discretization()
boucon = FEM.SetupModel()
solution = FEM.PostProcessing()
young = 1*10**9
poisson = 0.3
thick = 0.01
material = young/(1-poisson**2)*np.array(([[1,poisson,0],
                                    [poisson,1,0],
                                    [0,0,(1-poisson)/2]]))*thick
k = (3 - poisson)/(1 + poisson)
```

Define a grid

```python
lx, ly = 1, 1 # lx = b; ly = L
nx, ny = 8, 8 # mesh size
dc = 0.002 # amplitude of initial fracture 
p, t, nodaro, tipcra, nodcra, nodmid = sm.edgecrack(lx, ly, nx, ny, dc)
numtip = len(tipcra)
crapro = np.array(([True]))
rm = Grid.ReMesh( p, t, tipcra, nodcra, nodaro, dc, 0 )
```

## Main program

```python
N = 5
for step in range(N):
    p, t, moucra, roscra = rm.quaterelement( )
    p6, t6, qpe = mt.t3tot6(p,t,tipcra)
    topind = af.p2index(p6,nodaro[[3,2],:],1) 
    botind = af.p2index(p6,nodaro[[0,1],:],1)  
    rigind = af.p2index(p6,nodaro[[1,2],:],1)  
    midind = af.p2index(p6,nodmid[[0,-1],:],1)
    
    dirdof = np.concatenate((rigind*2, midind*2+1))
    dirval = 0*dirdof
    
    sig = 100000
    neudoft = np.zeros((np.int((len(topind) - 1)/2),3), np.int32)
    neuvalt = np.zeros((np.int((len(topind) - 1)/2),2))
    for i in range(np.int((len(topind) - 1)/2)):
        index = [2*i, 2*i + 1, 2*i + 2]
        neudoft[i,:] = topind[index]
        neuvalt[i,:] = [0, sig]
    
    neudofb = np.zeros((np.int((len(botind) - 1)/2),3), np.int32)
    neuvalb = np.zeros((np.int((len(botind) - 1)/2),2))
    for i in range(np.int((len(botind) - 1)/2)):
        index = [2*i, 2*i + 1, 2*i + 2]
        neudofb[i,:] = botind[index]
        neuvalb[i,:] = [0, -sig]
    
    neudof = np.concatenate((neudoft,neudofb), axis=0)
    neuval = np.concatenate((neuvalt,neuvalb), axis=0)

    lefhs = discre.stiffness( p6, t6, material, qpe) 
    righs = discre.loadsegment(p6, neudof, neuval)
    disp = solution.solver(lefhs,righs,dirdof,dirval)
    dispx = disp[::2]
    dispy = disp[1::2]
    U = np.concatenate((dispx, dispy), axis = 1)
    Ki, keq, craang = solution.SIF(p6, t6, numtip, disp, young, poisson, k, qpe )
    Ene = solution.energy( p6, t6, disp, material, qpe)
    if step < N:
        Sigxxi,Sigyyi,Sigxyi,error = solution.stresses(p6, t6, disp, material, qpe)
        parame = [Ene*young/(sig**2*lx**2)*thick, Ki/(sig*np.sqrt(math.pi*lx/2))*thick, max(error)[0]]
        fig, grid = plt.subplots()
        mp.trisurf(fig, grid, t6, p6 + U, 6, 0, parame)
        p, t = rm.refinement( crapro, error, 0, 'based on error' )
```
